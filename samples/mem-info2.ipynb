```python
import re
from typing import Optional


def parse_from_log(log_text: list[str], mem_info: Optional[MemoryInfoList] = None, show_free=False):
    # setup terminators
    RE_MEM_INFO_START = r"\*\*\* Memory info (.*) \*\*\*"
    MEM_INFO_END = "*********************"

    if not mem_info:
        mem_info = MemoryInfoList(show_free=show_free)
    else:
        mem_info.show_free = show_free

    # init
    in_mem_info = False
    nr = 0
    mem_info_log = []
    # find the meory_info lines in the (console) log output
    while nr < len(log_text):
        # if the regex matches, start a new map
        if match := re.match(RE_MEM_INFO_START, log_text[nr]):
            # start a new map
            mem_info_log = []
            in_mem_info = True
            # get the name of the map
            map_name = match[1]
        if in_mem_info:
            if log_text[nr].startswith(MEM_INFO_END):
                # At end of the map, add it to the memory info object
                in_mem_info = False
                if len(mem_info_log) > 0:
                    mem_info.append(mem_info_log, name=map_name)
            else:
                # add a line to the map
                mem_info_log.append(log_text[nr])
        nr += 1
    return mem_info


mem_info = parse_from_log(log_text)
```

```python
# just the first few
import time


def play(mem_info, start, end, delay=0.1, did=None):
    mem_info.diff_with = ()
    if not did:
        did = display(mem_info, display_id=True)
    else:
        update_display(mem_info, display_id=did.display_id)
    time.sleep(delay)
    for n in range(start, min(len(mem_info.memory_maps), end)):
        mem_info.diff_with = (n - 1, n)
        update_display(mem_info, display_id=did.display_id)
        time.sleep(delay)


play(mem_info, 0, 5, 2)
```
