```python
# %%micropython

# The following enables any function or method to be timed by adding an @timed_function decorator:
import time


def timed_function(f, *args, **kwargs):
    "enables any function or method to be timed by adding an @timed_function decorator"
    myname = str(f).split(" ")[1]

    def new_func(*args, **kwargs):
        t = time.ticks_us()
        result = f(*args, **kwargs)
        delta = time.ticks_diff(time.ticks_us(), t)
        print("Function `{}` WallTime = {:6.3f}ms".format(myname, delta / 1000))
        return result

    return new_func
```

```python
# %%micropython
import time
import random

@timed_function
def micro_slough(x:int):
    print("Zzzzzz...")
    time.sleep_ms(x)
    return time.time_ns()


for i in range(10):
    if i == 0:
        x = 0
    else:
        x = random.randint(10, 500)
    print(f"    {micro_slough(x):>19}")
```
